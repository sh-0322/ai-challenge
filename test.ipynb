```python
# 이 셀은 LLM API를 호출하여 스케쥴의 핵심 카테고리를 분석하는 함수를 정의합니다.
from openai import OpenAI

# --- LLM API Key 설정 ---
# 발급받은 OpenAI API Key를 여기에 붙여넣으세요.
# (실제 서비스에서는 이 키를 코드에 직접 노출하면 안 되지만, 지금은 테스트 목적입니다.)
#client = OpenAI(api_key = "ㅌ")

# -----------------------------------------------------------

# --- AI의 핵심 두뇌 함수 (최종 표준 버전) ---
def get_recommendation_final(schedule_name, knowledge_df):
    
    # --- 백엔드와 약속된 '정답지' (카드 혜택 카테고리 목록) ---
    BENEFIT_CATEGORIES = [
        "ALL_MERCHANTS", "TRANSPORTATION", "FUEL", "TELEPHONE_CHARGE", "MART_CONVENIENCE_STORE", 
        "ONLINE_SHOPPING", "SHOPPING", "FOOD", "CAFE_DESSERT", "BEAUTY_FITNESS", 
        "NO_SPENDING_REQUIREMENT", "UTILITY_BILLS_RENTAL", "HOSPITAL_PHARMACY", "PETS", 
        "EDUCATION_CHILDCARE", "AUTOMOBILE_HI_PASS", "LEISURE_SPORTS", "DIGITAL_SUBSCRIPTION", 
        "MOVIES_CULTURE", "EASY_PAYMENT", "AIRLINE_MILEAGE", "AIRPORT_LOUNGE_PRIORITY_PASS", 
        "PREMIUM", "TRAVEL_ACCOMMODATION", "OVERSEAS", "BUSINESS", "ETC"
    ]
    
    prompt = f"""
        당신은 사용자의 캘린더 일정을 분석하여 가장 관련 있는 소비 카테고리를 찾아주는 AI입니다.
        아래 "카테고리 목록" 중에서 "{schedule_name}"라는 일정 내용과 가장 관련 깊은 것 단 하나만 골라, 
        다른 설명은 일절 없이, 반드시 영어 카테고리 이름으로만 대답해주세요.
        만약 관련 있는 카테고리가 없다면 'ETC'라고 대답하세요.

        --- 카테고리 목록 ---
        {', '.join(BENEFIT_CATEGORIES)}
    """
    try:
        # 표준 v1.0.0 방식의 API 호출
        response = client.chat.completions.create(
            model="gpt-4",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.0
        )
        llm_category = response.choices[0].message.content.strip()
        if llm_category not in BENEFIT_CATEGORIES: llm_category = "ETC"
    except Exception as e:
        print(f"LLM 분석 실패: {e}")
        llm_category = "ETC"

    if llm_category != "ETC":
        result_df = knowledge_df[knowledge_df['benefit_category'] == llm_category]
        if not result_df.empty:
            top_card = result_df.iloc[0]
            return f"▶ '{llm_category}' 혜택! [{top_card['card_name']}] 카드의 '{top_card['benefit_info']}' 혜택을 추천합니다!"
        else:
            return f"'{llm_category}' 카테고리의 혜택은 있으나, 현재 보유한 지식 내에 적합한 카드가 없습니다."
    return "연관된 카드 혜택을 찾지 못했습니다."
```

```python
print("===== 🤖 [최종 버전] AI 캘린더 에이전트 추천 시뮬레이션 시작 🤖 =====")

# 1단계에서 로드한 모든 스케쥴을 하나씩 AI에게 물어봅니다.
for index, schedule in df_schedule.iterrows():
    schedule_name = schedule['schedule_name']
    
    # 2단계에서 만든 AI 두뇌 함수를 호출하여 추천 결과를 받습니다.
    recommendation = get_recommendation_final(schedule_name, df_card_benefits)
    
    # 결과를 출력합니다.
    print(f"\n🗓️ 입력된 일정: {schedule_name}")
    print(f"💬 AI의 추천: {recommendation}")
    print("-" * 60)

print("\n===== ✅ 시뮬레이션이 성공적으로 완료되었습니다. =====")
```
