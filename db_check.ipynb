```python
import mysql.connector
import pandas as pd

# 백엔드 개발자님께 받은 접속 정보를 여기에 입력하세요.
'''db_config = {
    'host': "54.66.228.99",
    'user': "root",
    'password': "ai-challenge-1234",
    'database': "ai_challenge",
    'port': 3306 # 받으신 포트 번호가 다르다면 수정해주세요.
}
'''

def load_data_from_ec2_mysql():
    connection = None
    try:
        print("DB에 다시 연결해서 테이블 목록을 확인합니다...")
        connection = mysql.connector.connect(**db_config)
        cursor = connection.cursor()

        query = "SHOW TABLES;"
        cursor.execute(query)

        tables = cursor.fetchall()

        print("\n--- 실제 데이터베이스에 있는 테이블 목록 ---")
        for table in tables:
            print(f"-> {table[0]}")
        print("-------------------------------------------")

    except mysql.connector.Error as e:
        print(f"에러 발생: {e}")
    finally:
        if connection and connection.is_connected():
            cursor.close()
            connection.close()
            print("정찰 완료. DB 연결을 닫았습니다.")
```

```python
# 위에서 만든 함수를 실행해서 데이터를 df 라는 변수에 담습니다.
df_card_benefits = load_data_from_ec2_mysql()

# Jupyter Notebook에서는 변수 이름만 마지막 줄에 써주면
# 표 형태로 예쁘게 결과를 보여줍니다.
df_card_benefits
```
